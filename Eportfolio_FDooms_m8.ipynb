Phase portraits and continuous relative phase (CRP) analysis are commonly used tools in the study of dynamical systems, particularly in the context of coordination and synchronization phenomena.

Phase Portrait: A phase portrait is a graphical representation of the behavior of a dynamical system in terms of its state variables. It typically plots one state variable against another, showing how the system's state evolves over time. Phase portraits can reveal information about the system's stability, attractors, and dynamics.

Continuous Relative Phase (CRP): CRP analysis is used to quantify the phase relationship between two oscillatory signals. It provides a measure of the relative timing or synchronization between the two signals. CRP analysis can reveal patterns of coordination or synchronization between different components or agents within a complex system.

In this Module we will use dummy variables.

```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example oscillatory signals
t = np.linspace(0, 10, 1000)  
omega1 = 2 * np.pi * 1  
omega2 = 2 * np.pi * 1.2  
signal1 = np.sin(omega1 * t)  
signal2 = np.sin(omega2 * t + np.pi/4)  

# Plot the phase portrait
plt.figure(figsize=(8, 6))
plt.plot(signal1, signal2, color='b')
plt.xlabel('Signal 1')
plt.ylabel('Signal 2')
plt.title('Phase Portrait')
plt.grid(True)
plt.show()

# Compute the continuous relative phase (CRP)
def compute_crp(signal1, signal2):
    phase1 = np.arctan2(np.imag(signal1), np.real(signal1))
    phase2 = np.arctan2(np.imag(signal2), np.real(signal2))
    crp = np.mod(phase2 - phase1 + np.pi, 2 * np.pi) - np.pi
    return crp

# Compute CRP between the two signals
crp = compute_crp(signal1, signal2)

# Plot the CRP
plt.figure(figsize=(8, 4))
plt.plot(t, crp, color='r')
plt.xlabel('Time')
plt.ylabel('Continuous Relative Phase (CRP)')
plt.title('CRP between Signal 1 and Signal 2')
plt.grid(True)
plt.show()
```

## Results

In the phase portrait, each point represents the state of the system at a particular time, with one coordinate representing the value of signal1 and the other representing the value of signal2.
Patterns in the phase portrait can reveal information about the relationship between the two signals. In this case it shwos that signal1 and signal2 are strongly coupled or synchronized as the datapoints are not scattered or clustered.

The CRP represents the phase difference between signal1 and signal2 as a function of time.
A CRP close to zero indicates in-phase synchronization, where the two signals are oscillating in sync. In this case we see that it ranges from -3 to 0. We can conclude that there is synchronization. However, we do see quite some movement between the points but as the it keeps going up and down in the same window we can conclude that there is no changes in the synchronization of the relationship.

## Intro to complexity
Complexity refers to the degree of intricacy or complication within a system, often involving numerous interacting components or variables. In a complex system, the behavior of the whole is not simply the sum of its parts; instead, it arises from the interactions and relationships between those parts.

Defining the level of complexity of a system involves assessing factors such as the number of components, dynamics and interactions, the presence of emergence, and the degree of unpredictability in system behavior. Emergence can be characterized by systems that feature interaction-dominant dynamics as opposed to component-dominant dynamics
In addition, many linear methods can be applied to study complex systems, and complex systems can exhibit (locally) linear behavior

## Tools for complex systems

Tools for understanding complex systems can vary depending on the nature of the system being studied, approaches we will see include:


Order changes: We will investigate order changes in the data to investigate the temporal signiture of the timeseries.

Topological data analysis (TDA) is a method for understanding the shape and structure of data. In the context of time series data, TDA can help identify persistent features such as loops, clusters, and other structures that might be indicative of the underlying dynamics.

Recurrence: Recurrence is the degree to which a system comes to inhabit the same or a very similar state. We can use a recurrence time series plot, this is a time series plotted against itself (auto-recurrence), with a point being placed on the plot each time a prior state is exhibited by the system. We will investigate plots and use statisticall analysis to gain a better unerstanding of the underlying structure.

EDM: Empirical Dynamic Modeling (EDM) is a data-driven approach used in the field of complex systems analysis to understand and predict the behavior of dynamical systems. It's particularly useful when dealing with nonlinear and complex systems where traditional mathematical modeling techniques may be insufficient.
